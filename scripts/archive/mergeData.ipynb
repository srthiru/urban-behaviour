```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import glob
import os
import numpy as np
import math
import datetime
import json
```

```python
path = 'taxi data/december/green_tripdata_2018-12.csv'
greendf = pd.read_csv(path)


path = 'taxi data/december/fhv_tripdata_2018-12.csv'
forhiredf = pd.read_csv(path)


path = 'taxi data/december/yellow_tripdata_2018-12.csv'
yellowdf = pd.read_csv(path)

```

```python
green = greendf[["lpep_pickup_datetime", "lpep_dropoff_datetime", 
           "PULocationID", "DOLocationID"]].copy()

print(green.shape)

green = green.dropna(how='any')
green = green.reset_index(drop=True)
green = green.rename(columns={'lpep_pickup_datetime': 'pickup', 
                              'lpep_dropoff_datetime': 'dropoff',
                              'PULocationID':'pickupLocID',
                              'DOLocationID':'dropoffLocID'                              
                             })
print(green.shape)

green.head()
```

```python
forhire = forhiredf[["Pickup_DateTime", "DropOff_datetime", 
           "PUlocationID", "DOlocationID"]].copy()

print(forhire.shape)

forhire = forhire.dropna(how='any')
forhire = forhire.reset_index(drop=True)
forhire = forhire.rename(columns={'Pickup_DateTime': 'pickup', 
                              'DropOff_datetime': 'dropoff',
                              'PUlocationID':'pickupLocID',
                              'DOlocationID':'dropoffLocID'                              
                             })
print(forhire.shape)

forhire.head()
```

```python
yellow = yellowdf[["tpep_pickup_datetime", "tpep_dropoff_datetime", 
           "PULocationID", "DOLocationID"]].copy()

print(yellow.shape)

yellow = yellow.dropna(how='any')
yellow = yellow.reset_index(drop=True)
yellow = yellow.rename(columns={'tpep_pickup_datetime': 'pickup', 
                              'tpep_dropoff_datetime': 'dropoff',
                              'PULocationID':'pickupLocID',
                              'DOLocationID':'dropoffLocID'                              
                             })
print(yellow.shape)

yellow.head()
```

```python
allData = pd.concat([green, forhire, yellow], ignore_index=True)
allData.head()
```

```python
allData["pickup"] = pd.to_datetime(allData['pickup'])
allData['pickup'] = allData['pickup'].dt.date

allData["dropoff"] = pd.to_datetime(allData['dropoff'])
allData['dropoff'] = allData['dropoff'].dt.date
# yellow.head()

allData.shape

allData.to_csv('taxi data/all data/december.csv', index=False)
```
