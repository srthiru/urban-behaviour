```python
import nltk
nltk.download('vader_lexicon')
```

```python

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

nltk_sentiment = tweets_df['text'].apply(lambda x: sia.polarity_scores(x))
```

```python
pos_sentiment = nltk_sentiment.apply(lambda x: x['pos'])
neg_sentiment = nltk_sentiment.apply(lambda x: x['neg'])
neu_sentiment = nltk_sentiment.apply(lambda x: x['neu'])
```

```python
def classify_sentiment(x):
    if x['pos'] > 0.2:
        return "Positive"
    if x['neg'] > 0.2:
        return "Negative"
    if x['neu'] > 0.2:
        return "Neutral"
    return "Mixed"

classification = nltk_sentiment.apply(lambda x: classify_sentiment(x))
```

```python
tweets_df['text'][classification == "Positive"]
```

```python
import string
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stop_words = stopwords.words('english')

def preprocess_text(text):
    text = text.lower()
    text = "".join([char for char in text if char not in string.punctuation])
    words = word_tokenize(text)
    filtered_words = [word for word in words if word not in stop_words]
    stemmed_words = [porter.stem(word) for word in filtered_words]
    return stemmed_words
    
```

```python
preprocessed = tweets_df['text'].apply(lambda x: preprocess_text(x))
```

```python
processed_df = pd.DataFrame()
processed_df['id'] = tweets_df['id']
processed_df['text'] = tweets_df['text']
processed_df['Sentiment'] = classification
processed_df['Extracted Keywords'] = preprocessed
processed_df['lang'] = tweets_df['lang']
processed_df['retweet_count'] = tweets_df['retweet_count']
processed_df['coordinates'] = tweets_df['coordinates']
```
